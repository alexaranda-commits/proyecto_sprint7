##1.  Cargar los datos.

```python
#inicializamos librerias
import pandas as pd
import plotly.graph_objects as go

#vaciamos los datos de  vehicles_us.csv en un dataframe
df_vehicles = pd.read_csv('../vehicles_us.csv')
#revisamos la informacion del dataframe
df_vehicles.info()
#mostramos las primeras 10 filas del dataframe
display(df_vehicles.head(10))
```

##Paso 2.  Corregir los datos

Despues de un analysis exploratorio de los datos, observamos que tenemos que corregir tipos de datos en las columnas:  
price -> int a float  
model_year -> float a int  
cylinders -> float a int  
is_4wd -> float a boolean  
date_posted -> object a datetime  

tendremos que decidir como manejar los nulos en las columnas:  
model_year, cambiar nulos a 0  
cylinders, cambiar nulos a 0  
odometer, cambiar nulos 0  
paint_color, cambiar nulos a unknown  
is_4wd, revisar si estan representados con 0 los autos sin 4wd, o si estan representados con NaN. determinar que hacer con nulos despues de eso  

```python
#Comenzamos corrigiendo los valores nulos
df_vehicles.fillna({'model_year' : 0}, inplace=True)
df_vehicles.fillna({'cylinders' : 0}, inplace=True)
df_vehicles.fillna({'odometer' : 0}, inplace=True)
df_vehicles.fillna({'paint_color' : 'unknown'}, inplace=True)
#Revisamos los valores unicos de la columna is_4wd
print(df_vehicles['is_4wd'].unique())
#Rellenamos los valores nulos de is_4wd con False
df_vehicles.fillna({'is_4wd' : 0}, inplace=True)
#Verificamos que no queden valores nulos
print(df_vehicles.isnull().sum())
```

```python
#cambiamos los tipos de datos de las columnas que lo requieren
df_vehicles['price'] = df_vehicles['price'].astype(float)
df_vehicles['model_year'] = df_vehicles['model_year'].astype(int)
df_vehicles['cylinders'] = df_vehicles['cylinders'].astype(int)
df_vehicles['date_posted'] = pd.to_datetime(df_vehicles['date_posted'])
#cambiamos el tipo de dato de is_4wd a booleano
df_vehicles['is_4wd'] = df_vehicles['is_4wd'].astype(bool)
#verificamos el cambio de tipo de dato
print(df_vehicles.dtypes)
display(df_vehicles.head(10))
```

Vamos a separar la columna model, en manufactuere y model name, para asi tener informacion mas robusta

```python
df_vehicles['manufacturer'] = df_vehicles['model'].apply(lambda x: x.split(' ')[0] if isinstance(x, str) else 'unknown')
df_vehicles['model_name'] = df_vehicles['model'].apply(lambda x: ' '.join(x.split(' ')[1:]) if isinstance(x, str) else 'unknown')
#verificamos el cambio en el dataframe
display(df_vehicles[['model', 'manufacturer', 'model_name']].head(10))
```

##Paso 3.  Analysis exploratorio de los datos.  

Crea graficos para los siguientes casos:  
Vehiculos disponibles por fabricante
Tipos de vehiculo por fabricante  
Distribucion de precios  
Grafico de caja de Precios  
Histograma de condicion vs año  
Grafico de dispercion de relacion entre odometro y precio

```python
#Vehiculos disponibles por fabricante
fig1 = go.Figure(data=[go.Bar(x=df_vehicles['manufacturer'].value_counts().index,
                              y=df_vehicles['manufacturer'].value_counts().values)],
                              layout=go.Layout(title='Vehiculos disponibles por fabricante',
                                  xaxis_title='Fabricante',
                                  yaxis_title='Cantidad de vehiculos'))

#mostramos el grafico
fig1.show()
```

```python
#Tipos de Vehiculos por fabricante
fig2 = go.Figure()
for manufacturer in df_vehicles['manufacturer'].unique():
    df_manufacturer = df_vehicles[df_vehicles['manufacturer'] == manufacturer]
    fig2.add_trace(go.Bar(x=df_manufacturer['type'].value_counts().index,
                          y=df_manufacturer['type'].value_counts().values,
                          name=manufacturer))
fig2.update_layout(title='Tipos de Vehiculos por fabricante',
                   xaxis_title='Tipo de Vehiculo',
                   yaxis_title='Cantidad de Vehiculos',
                   barmode='stack')
#mostramos el grafico
fig2.show()
```

```python
#Condicion vs fabricante
fig3 = go.Figure()
for condition in df_vehicles['condition'].unique():
    df_condition = df_vehicles[df_vehicles['condition'] == condition]
    fig3.add_trace(go.Histogram(x=df_condition['manufacturer'],
                                name=condition))
fig3.update_layout(title='Condicion vs Fabricante',
                   xaxis_title='Fabricante',
                   yaxis_title='Cantidad de Vehiculos')
                   
#mostramos el grafico
fig3.show()
```

```python
#distribucion de precios
fig4 = go.Figure(data=[go.Box(x=df_vehicles['price'])],
                                   layout=go.Layout(title='Distribucion de Precios',
                                   xaxis_title='Precio',
                                   yaxis_title='Cantidad de Vehiculos'))
#mostramos el grafico
fig4.show()
```

```python
#grafico de caja de precios por fabricante
fig5 = go.Figure()
for manufacturer in df_vehicles['manufacturer'].unique():
    df_manufacturer = df_vehicles[df_vehicles['manufacturer'] == manufacturer]
    fig5.add_trace(go.Box(y=df_manufacturer['price'],
                          name=manufacturer))
fig5.update_layout(title='Distribucion de precios entre fabricantes',
                   yaxis_title='Precio (USD)')
#mostramos el grafico
fig5.show()
```

```python
#Histograma de condicion vs año 
fig6 = go.Figure()
#Removemos valores de model_year iguales a 0
df_vehicles_no0 = df_vehicles[df_vehicles['model_year'] != 0]
for condition in df_vehicles_no0['condition'].unique():
    df_condition = df_vehicles_no0[df_vehicles_no0['condition'] == condition]
    fig6.add_trace(go.Histogram(x=df_condition['model_year'],
                                name=condition,
                                autobinx=True,
                                opacity=0.7))
fig6.update_layout(title='Condicion vs Año ',
                     xaxis_title='Año',
                     yaxis_title='Cantidad de Vehiculos',
                     barmode='overlay')
#mostramos el grafico
fig6.show()
```

```python
#Grafico de dispersion de relacion entre odometro y precio
fig7 = go.Figure()
fig7.add_trace(go.Scatter(x=df_vehicles['odometer'],
                            y=df_vehicles['price'],
                            mode='markers',
                            marker=dict(size=5, color='blue', opacity=0.5)))
fig7.update_layout(title='Relación entre Odómetro y Precio',
                     xaxis_title='Odómetro (millas)',
                     yaxis_title='Precio (USD)')
#mostramos el grafico
fig7.show()
```
